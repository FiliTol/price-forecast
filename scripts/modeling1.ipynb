```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
dic_listings = pd.read_pickle("../data/pickles/december_listings.pkl")
dic_listings = dic_listings.drop(["name"], axis=1)
dic_reviews = pd.read_csv("../data/2023dic/d_reviews.csv")
dic_reviews = dic_reviews[["listing_id", "id", "date", "reviewer_id", "comments"]]
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text

analyzer = SentimentIntensityAnalyzer()
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = scores['pos']
    return sentiment
```

```python
# Figure out the sentiment for the two descriptive features
dic_listings["neighborhood_overview"] = dic_listings["neighborhood_overview"].apply(preprocess_text)
dic_listings["neighborhood_overview"] = dic_listings["neighborhood_overview"].apply(get_sentiment)

dic_listings["host_about"] = dic_listings["host_about"].apply(preprocess_text)
dic_listings["host_about"] = dic_listings["host_about"].apply(get_sentiment)
```
